```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/covid_19_indonesia_time_series_all.csv')
df
```

```python
df["Location"]
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

Mengubah waktu.

```python
data = df.sort_values(by="Date", ascending=False).head(45)
data
```

Melihat data yang ada maka dilakukan penyesuaian menggunakan tanggal paling akhir sebagai acuan karena menggambarkan nilai keseluruhan yang ada pada tiap daerah.

```python
dell = data[data["Location"]=="Indonesia"].index
data.drop(dell, inplace=True)
```

Menghapus baris Indonesia karena bukan merupakan bagian data dari pulau yang diperlukan.

```python
totKasus = data[(data["Date"] == "2020-08-03")]
totKasus
```

Filtering berdasarkan tanggal yang diperlukan yaitu 03-08-2020.

```python
totKasus["Location"]
```

Menampilkan hasil filetring berisikan informasi lokasi.

```python
print ('Min: ' + str(totKasus["Total Active Cases"].min()))
print ('Mode: ' + str(totKasus["Total Active Cases"].mode()[0]))
print ('Median: ' + str(totKasus["Total Active Cases"].median()))
print ('Mean: ' + str(totKasus["Total Active Cases"].mean()))
print ('Max: ' + str(totKasus["Total Active Cases"].max()))
```

Menghitung central tendency secara manual.

```python
totKasus.describe()
```

Melakukan konfirmasi menggunakan command yang ada untuk membandingkan dengan nilai yang sudah ditemukan.

```python
totKasus[["Total Active Cases"]].cov()
```

Menghitung covariance

```python
covv=totKasus[["New Cases","New Deaths","Total Deaths","Total Recovered","Total Active Cases"]].corr()
covv
```

Menghitung corelasi

```python
sns.heatmap(covv)
```
